```python
import pandas as pd
import joblib
```

```python
# Load Model & Preprocessor
preprocessor = joblib.load('preprocessor_credit_risk.pkl')
logreg_model = joblib.load('model_logreg_credit_risk.pkl')
```

```python
# Data Nasabah Baru
data_nasabah_baru = {
    'loan_amnt': 15000,
    'int_rate': 11.5,
    'annual_inc': 60000,
    'dti': 15.5,
    'delinq_2yrs': 0,
    'inq_last_6mths': 1,
    'open_acc': 10,
    'pub_rec': 0,
    'revol_bal': 12000,
    'revol_util': 45.0,
    'total_acc': 20,
    'collections_12_mths_ex_med': 0,
    'acc_now_delinq': 0,
    'emp_length_num': 5,
    'term_num': 36,
    'grade': 'B',
    'home_ownership': 'RENT',
    'verification_status': 'Verified',
    'purpose': 'credit_card',
    'initial_list_status': 'f',
    
    # Menambahkan kolom yang ditagih oleh preprocessor:
    'installment': 494.0,           # Nilai estimasi cicilan
    'tot_cur_bal': 150000.0,        # Nilai estimasi saldo saat ini
    'total_rev_hi_lim': 30000.0,    # Nilai estimasi limit kartu
    'tot_coll_amt': 0.0             # Nilai estimasi tunggakan kolektor
}

df_nasabah = pd.DataFrame([data_nasabah_baru])

```

```python
# Jalankan Prediksi
try:
    nasabah_processed = preprocessor.transform(df_nasabah)
    prediksi = logreg_model.predict(nasabah_processed)[0]
    probabilitas = logreg_model.predict_proba(nasabah_processed)[0]

    print("="*30)
    print("HASIL ANALISIS KREDIT")
    if prediksi == 1:
        print(f"STATUS    : DISERUJUI (Good Loan)")
        print(f"KEYAKINAN : {probabilitas[1]*100:.2f}%")
    else:
        print(f"STATUS    : DITOLAK (Bad Loan / Beresiko)")
        print(f"KEYAKINAN : {probabilitas[0]*100:.2f}%")
    print("="*30)
except Exception as e:
    print(f"Masih terjadi error: {e}")
```
